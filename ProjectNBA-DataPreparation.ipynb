Load df_combined csv into a df

```python
# Provide full path to df_combined CSV file in Google Drive
df_combined = '/content/drive/MyDrive/COMP 333/NBA-Data/df_combined.csv'

# Load the CSV file into a DataFrame
df_combined = pd.read_csv(df_combined, encoding='latin', sep=',')
```

Remove any rows with null data in any columns

```python
df_combined_cleaned = df_combined.dropna()

df_combined_cleaned
```

Remove outliers. Remove any players who have played less than half a given season. Usual seasons are 82 games. Other seasons were cut short due to special circumstances.

```python
df_not_in_outlierYears = df_combined_cleaned[~df_combined_cleaned['Year'].isin(['2020-2021', '1998-1999', '2011-2012'])]
df_not_in_outlierYears_under41 = df_not_in_outlierYears[df_not_in_outlierYears['G'] < 41]
print(df_not_in_outlierYears_under41)

df_year_2020_2021 = df_combined_cleaned[df_combined_cleaned.isin(['2020-2021'])]
df_year_2020_2021_under36 = df_combined_cleaned[df_combined_cleaned['G'] < 36]

df_year_1998_1999 = df_combined_cleaned[df_combined_cleaned.isin(['1998-1999'])]
df_year_1998_1999_under25 = df_combined_cleaned[df_combined_cleaned['G'] < 25]

df_year_2011_2012 = df_combined_cleaned[df_combined_cleaned.isin(['2011-2012'])]
df_year_2011_2012_under33 = df_combined_cleaned[df_combined_cleaned['G'] < 33]

df_combined_cleaned = pd.merge(df_combined_cleaned, df_not_in_outlierYears_under41, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
df_combined_cleaned = pd.merge(df_combined_cleaned, df_year_2020_2021_under36, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
df_combined_cleaned = pd.merge(df_combined_cleaned, df_year_1998_1999_under25, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
df_combined_cleaned = pd.merge(df_combined_cleaned, df_year_2011_2012_under33, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)

print(df_combined_cleaned)
```
